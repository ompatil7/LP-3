Classify the email using the binary classification method. Email Spam detection has two states: 

a) Normal State – Not Spam, 
b) Abnormal State – Spam. 

Use K-Nearest Neighbors and Support Vector Machine for classification. Analyze their performance.

Dataset link: The emails.csv dataset on the Kaggle
https://www.kaggle.com/datasets/balaka18/email-spam-classification-dataset-csv

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("emails.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# splitting the data set
x = df.drop(["Prediction","Email No."],axis=1)
y = df["Prediction"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# K-Nearest Neighbour 
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

# print("K-Nearest Neighbors Classification Report")
# print(classification_report(y_test, y_pred_knn))
print("K-Nearest Neighbors Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("\n")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("\n")


# ye graph print karega confusion matrix ka, if u want u can add this will be good
import seaborn as sns
import matplotlib.pyplot as plt
cm_knn = confusion_matrix(y_test, y_pred_knn)

# Confusion matrix for KNN
cm_knn = confusion_matrix(y_test, y_pred_knn)
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues', annot_kws={'size': 14}, vmin=0, vmax=cm_knn.max())
plt.title('KNN Confusion Matrix')
plt.show()
```

```python
# Support Vector Machine (SVM)
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

print("Support Vector Machine Classification Report")
print(classification_report(y_test, y_pred_svm))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))

# ye graph print karega confusion matrix ka, if u want u can add this will be good
import seaborn as sns
import matplotlib.pyplot as plt
# Confusion matrix for SVM
cm_svm = confusion_matrix(y_test, y_pred_svm)
sns.heatmap(cm_svm, annot=True, fmt='d')
plt.title('SVM Confusion Matrix')
plt.show()
```

```python
acc_knn = accuracy_score(y_test, y_pred_knn)
acc_svm = accuracy_score(y_test, y_pred_svm)
print(f"KNN Accuracy: {acc_knn * 100:.2f}%")
print(f"SVM Accuracy: {acc_svm * 100:.2f}%")
```

## Theory



## K-Nearest Neighbors (KNN):
Classifies an email based on the classes of its "k" nearest neighbors in the feature space. It is sensitive to noisy data but works well with a small number of dimensions.
## Support Vector Machine (SVM): 
SVM finds a decision boundary that maximally separates spam from not spam, focusing on the most critical points. It is effective in high-dimensional spaces and handles text data well.


------------------------------------------------------------------------------------------------
## Evaluation Metrics
Accuracy: The proportion of correctly classified emails (either spam or not spam) out of the total emails.
Classification Report: Displays precision, recall, and F1-score, helping in analyzing the model's performance on each class.
Confusion Matrix: Shows the true positives, true negatives, false positives, and false negatives, providing insights into where the model may be misclassifying.


------------------------------------------------------------------------------------------------
True Negative (TN) = 645
False Positive (FP) = 94
False Negative (FN) = 48
True Positive (TP) = 248

Formula:
Precision (also called Positive Predictive Value) measures the accuracy of positive predictions.
Precision = TP / TP + FP

Recall (also called Sensitivity or True Positive Rate) measures the model's ability to correctly identify all positive instances.
Recall= TP / FN + TP
​
The F1-score is the harmonic mean of Precision and Recall. It balances the two metrics, especially useful when there’s an uneven class distribution.
F1-Score=2× (Precision*Recall / Precision+Recall )
